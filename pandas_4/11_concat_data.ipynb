```python
df2 = pd.read_csv('files/mpg.csv')
```

```python
df.head()
```

# another example

```python
data_1 = {'A': ['A0','A1','A2','A3','A4'], 'B': ['B0','B1','B2','B3','B4']}
```

```python
df1 = pd.DataFrame(data_1)
```

```python
df1
```

```python
data_2 = {'C': ['C0','C1','C2','C3','C4'], 'D': ['D0','D1','D2','D3','D4']}
```

```python
df2 = pd.DataFrame(data_2)
```

```python
df2
```

```python
# joins
pd.concat([df1, df2], axis = 1)
```

```python
# union
pd.concat([df1, df2], axis = 0)
```

```python
# match the column names for proper union
df2.columns = df1.columns
```

```python
pd.concat([df1, df2])
```
